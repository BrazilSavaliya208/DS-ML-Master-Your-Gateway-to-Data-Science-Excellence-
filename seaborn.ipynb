## 2). dist plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm 
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['size'])
plt.show()
```

```python
sns.distplot(tips_df['tip'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'])
plt.show()
```

```python


sns.distplot(tips_df['total_bill'],bins=100)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],hist=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],kde=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],rug=True)
plt.show()
```

```python

sns.distplot(tips_df['total_bill'],fit=norm)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],fit=norm,kde=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],hist_kws={'color':'y','edgecolor':'r','linewidth':3,'linestyle':'--','alpha':0.9})
```

```python
sns.distplot(tips_df['total_bill'],hist_kws={'color':'y','edgecolor':'r','linewidth':3,'linestyle':'--','alpha':0.9},kde_kws={'color':'g','linestyle':'--','linewidth':8,'alpha':0.9})
```

```python
sns.distplot(tips_df['total_bill'],rug=True, rug_kws={'color':'#0426d0'
                                                      ,'linewidth':5,'linestyle':'--','alpha':0.9})
```

```python
sns.distplot(tips_df['total_bill'],fit_kws = {'color':'#8e00ce', 'linewidth':12, 'linestyle':'--', 'alpha':0.4},fit=norm)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],color='k')
```

```python
sns.distplot(tips_df['total_bill'],vertical=True)
```

```python
sns.distplot(tips_df['total_bill'],norm_hist=True)
```

```python
# axlabel = x label name change

sns.distplot(tips_df['total_bill'],axlabel='Total bill')
```

```python
sns.distplot(tips_df['total_bill'],label='Hello')
plt.legend()
```

```python
sns.distplot(tips_df['total_bill'],label='total_bill')
sns.distplot(tips_df['size'],label='size')
sns.distplot(tips_df['tip'],label='tip')
plt.legend()
```

```python
plt.figure(figsize=(16,9))

a={'fontsize':10,
  'fontstyle':'italic',
  'color':'k',
  'alpha':0.9,
  'rotation':'vertical',
  'verticalalignment':'center',
  'backgroundcolor':'w'}

sns.heatmap(globalWarming_df,cmap='coolwarm',annot=True,annot_kws=a)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df,cmap='coolwarm',annot=True,linewidth=4)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df,cbar=False,xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(16,9))


a={'orientation':'horizontal',
  'shrink':1,
  'extend':'min',
  'extendfrac':0.1,
  'ticks':np.arange(0,22),
  'drawedges':True}

sns.heatmap(globalWarming_df,cbar_kws=a)
```

```python
plt.figure(figsize=(16,9))
    
a = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
sns.heatmap(globalWarming_df, xticklabels=np.arange(0,15),yticklabels = a)
```

```python
plt.figure(figsize=(16,9))
 
ax=sns.heatmap(globalWarming_df)
ax.set(title="Graph",xlabel='Number',ylabel='String')
sns.set(font_scale=8)
```
